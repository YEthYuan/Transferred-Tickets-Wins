```python
import matplotlib.pyplot as plt
import numpy as np
import math

font_mid = 12
font_small = 10
font_board = 1.5
lw = 3
```

```python
lw = 3
rst = [28.7,33.17,33.7,34.21,34.7,34.33,34.66]
ft_inher = [31.01,34.37,34.61,35.14,34.7,35.33,35.89]
ft_reinit = [28.57,31.12,32.57,33.16,34.3,34.79,34.63]


x_overall = [0.01,0.03,0.05,0.07,0.1,0.15,0.2]

# rst_N = [55.38,68.38,71.01,73.7,74.63,76.09,77.41,76.14,77.58,78.3,75.9,79.7]
# ft_inher_N = [67.92,72.69,75.28,76.93,78.24,79.27,78.83,79.42,80.14,81.08,80.34,81.29]
# ft_reinit_N = [61.6,69.12,71.09,72.45,73.11,74.35,74.79,75.74,75.09,75.46,77.16,78.92]

fig, ax1 = plt.subplots()
ax1.set_xlabel(r"$\bf{Remaining \ Ratio}$", fontsize = 16)
ax1.set_ylabel(r"$\bf{Robust \ Accuracy \ (\%)}$", fontsize = 16)

ax1.plot(x_overall, rst, c='tab:red', marker='s', markersize=3*lw, label=r"$\bf{RST}$", lw=lw)
ax1.plot(x_overall, ft_inher, c='tab:green', marker='^', markersize=3*lw, label=r"$\bf{FT\_RST\_Inherit}$", lw=lw)
ax1.plot(x_overall, ft_reinit, c='tab:blue', marker='o', markersize=3*lw, label=r"$\bf{FT\_RST\_Reinit}$", lw=lw)

# ax1.plot(x_overall, rst_N, c='tab:orange', marker='p', markersize=3*lw, label=r"$\bf{RST@Natural}$", lw=lw)
# ax1.plot(x_overall, ft_inher_N, c='tab:purple', marker='*', markersize=3*lw, label=r"$\bf{FT-RST-Inherit@Natural}$", lw=lw)
# ax1.plot(x_overall, ft_reinit_N, c='tab:gray', marker='D', markersize=3*lw, label=r"$\bf{FT-RST-Reinit@Natural}$", lw=lw)

ax1.tick_params(axis = 'both', which = 'major', labelsize = 14)

# ax1.set_ylim([35, 50])
xticks = np.arange(0, 0.21, 0.05)
ax1.set_xticks(xticks)
ax1.set_xlim([0, 0.21])
# yticks = np.arange(35, 51, 2)
# ax1.set_ylim([34, 83])
# ax1.set_yticks(yticks)

ax1.spines['bottom'].set_linewidth(font_board)
ax1.spines['bottom'].set_color('black')
ax1.spines['left'].set_linewidth(font_board)
ax1.spines['left'].set_color('black')
ax1.spines['top'].set_linewidth(font_board)
ax1.spines['top'].set_color('black')
ax1.spines['right'].set_linewidth(font_board)
ax1.spines['right'].set_color('black')

leg = ax1.legend(fontsize=font_mid, loc='best', ncol=1)
leg.get_frame().set_edgecolor("black")
leg.get_frame().set_linewidth(3)
ax1.grid(axis='both', color='gray', linestyle='-', linewidth=0.3)


fig.tight_layout()  # otherwise the right y-label is slightly clipped
# ax1.grid(True)
plt.savefig("ResNet18@Cifar10@RST_EPS12_sub.svg")
```
