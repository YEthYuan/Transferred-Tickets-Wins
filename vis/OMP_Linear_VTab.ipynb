```python
# Make Legend
def export_legend(ax, filename="Legend.svg"):
    fig2 = plt.figure(figsize=(8,1))
    ax2 = fig2.add_subplot()
    ax2.axis('off')

    legend = ax2.legend(*ax.get_legend_handles_labels(), frameon=False, loc='lower center', fontsize=font_mid, ncol=4,)
    
    fig  = legend.figure

    fig.canvas.draw()
    bbox  = legend.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    fig.savefig(filename, dpi=100, bbox_inches=bbox)
```
