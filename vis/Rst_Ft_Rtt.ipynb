```python
import matplotlib.pyplot as plt
import numpy as np
import math

font_mid = 12
font_small = 10
font_board = 1.5
lw = 3
```

```python
lw = 2.5
rst = [37.59,43.96,45.81,46.97,47,47.2,47.71,46.74,45.73,38.5]
ft_inher = [41.11,45.98,46.95,48.19,47.42,48.42,47.49,48.87,49.55,49.96]
ft_reinit = [37,45.3,47.35,48.34,47.83,49.07,50.05,50.81,50.83,50.91]

x_overall = [0.01,0.03,0.05,0.07,0.1,0.15,0.2,0.4,0.6,0.8]

# rst_N = [55.38,71.01,74.63,77.41,78.3,75.9,79.7,79.34,77.68,66.4]
# ft_inher_N = [67.92,75.28,78.24,78.83,81.08,80.34,81.29,81.86,80.62,81.56]
# ft_reinit_N = [61.92,75.54,76.56,76.87,78.57,80.29,80.32,81.81,82.44,81.99]

fig, ax1 = plt.subplots()
ax1.set_xlabel(r"$\bf{Remaining  \ Ratio}$", fontsize = 16)
ax1.set_ylabel(r"$\bf{Robust \ Accuracy \ (\%)}$", fontsize = 16)

ax1.plot(x_overall, rst, c='tab:red', marker='s', markersize=3*lw, label=r"$\bf{RST}$", lw=lw)
ax1.plot(x_overall, ft_inher, c='tab:green', marker='^', markersize=3*lw, label=r"$\bf{FT\_RST\_Inherit}$", lw=lw)
ax1.plot(x_overall, ft_reinit, c='tab:blue', marker='o', markersize=3*lw, label=r"$\bf{Adv\_RTT}$", lw=lw)

# ax1.plot(x_overall, rst_N, c='tab:red', marker='p', markersize=3*lw, label=r"$\bf{RST@Natural}$", lw=lw)
# ax1.plot(x_overall, ft_inher_N, c='tab:green', marker='*', markersize=3*lw, label=r"$\bf{FT-RST-Inherit@Natural}$", lw=lw)
# ax1.plot(x_overall, ft_reinit_N, c='tab:blue', marker='D', markersize=3*lw, label=r"$\bf{Adv\_RTT@Natural}$", lw=lw)

ax1.tick_params(axis = 'both', which = 'major', labelsize = 14)

# ax1.set_ylim([67, 78])
# ax1.set_xlim([-0.5,4.5])

ax1.spines['bottom'].set_linewidth(font_board)
ax1.spines['bottom'].set_color('black')
ax1.spines['left'].set_linewidth(font_board)
ax1.spines['left'].set_color('black')
ax1.spines['top'].set_linewidth(font_board)
ax1.spines['top'].set_color('black')
ax1.spines['right'].set_linewidth(font_board)
ax1.spines['right'].set_color('black')

leg = ax1.legend(fontsize=font_mid, loc='best')
leg.get_frame().set_edgecolor("black")
leg.get_frame().set_linewidth(3)
ax1.grid(axis='both', color='gray', linestyle='-', linewidth=0.3)


fig.tight_layout()  # otherwise the right y-label is slightly clipped
# ax1.grid(True)
plt.savefig("ResNet18@Cifar10@RST_RTT.svg")
```
